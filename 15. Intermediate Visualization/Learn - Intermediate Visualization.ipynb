# Axes

```python
# create some data to use for the plot
dt = 0.001
t = np.arange(0.0, 10.0, dt)
r = np.exp(-t[:1000] / 0.05)  # impulse response
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)] * dt  # colored noise

fig, main_ax = plt.subplots()
main_ax.plot(t, s)
main_ax.set_xlim(0, 1)
main_ax.set_ylim(1.1 * np.min(s), 2 * np.max(s))
main_ax.set_xlabel('time (s)')
main_ax.set_ylabel('current (nA)')
main_ax.set_title('Gaussian colored noise')

# this is an inset axes over the main axes
right_inset_ax = fig.add_axes([.65, .6, .2, .2], facecolor='k') #  [left percentage, bottom percentage, width, height]
right_inset_ax.hist(s, 400, density=True)
right_inset_ax.set(title='Probability', xticks=[], yticks=[])

# this is another inset axes over the main axes
left_inset_ax = fig.add_axes([.2, .6, .2, .2], facecolor='k') #  [left percentage, bottom percentage, width, height]
left_inset_ax.plot(t[:len(r)], r)
left_inset_ax.set(title='Impulse response', xlim=(0, .2), xticks=[], yticks=[])

plt.show()
```

# Annotations

## Straight Arrow

```python
fig, ax = plt.subplots()
 
t = np.arange(0.0, 1.0, 0.001)
s = np.sin(2 * np.pi * t)
line = ax.plot(t, s, lw=2)
 
ax.annotate('Peak Value', xy=(.25, 1),
                xytext=(1, 1),arrowprops=dict(facecolor='yellow',
                                shrink=0.05),xycoords="data",)
 
ax.set_ylim(-1.5, 1.5)
plt.show()
```

## Curved Arrow

```python
fig, ax = plt.subplots()

x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle3,angleA=0,angleB=-90"))

plt.show()
```

## Datapoint Label

```python
y = [3.2, 3.9, 3.7, 3.5, 3.02199]
x = [0.15, 0.3, 0.45, 0.6, 0.75]
n = [155, "outlier", 293, 230, 670]
 
fig, ax = plt.subplots()
ax.scatter(x, y)
 
for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))
    
plt.show()
```

## More Complicated

```python
labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
shop_a = [20, 33, 30, 28, 27]
shop_b = [25, 32, 33, 20, 25]
 
x = np.arange(len(labels))  
width = 0.35  # the width of the bars
 
fig, ax = plt.subplots(figsize=(12,6))
rects1 = ax.bar(x - width/2, shop_a, width, label='Sales-a')
rects2 = ax.bar(x + width/2, shop_b, width, label='sales-b')
 
 
ax.set_ylabel('Sales')
ax.set_title('Sales report of 2 shops')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim([None, max(max(shop_a), max(shop_b))*1.2])
ax.legend()
 
 
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points", size=16,color="Green",
                    ha='center', va='bottom')
 
 
autolabel(rects1)
autolabel(rects2)

# save to Google Drive
plt.savefig("/content/drive/MyDrive/test.png")

fig.tight_layout()
plt.show()
```
