### Solution - Task 1
Creating a dataframe using lists

```python
import pandas as pd
import numpy as np
```

```python
# column name list
column_name = ['ID', 'Gender', 'Ever_Married',
               'Age', 'Graduated', 'Profession',
               'Work_Experiences', 'Spending_Score', 'Family_Size',
               'Var_1', 'Segementation']
```

```python
# creating a list for each column
list_id = [462809, 462643, 466315, 461735]
list_gender = ['Male', 'Female', 'Female', 'Male']
list_mariage = ['No', 'Yes', 'Yes', 'Yes']
list_age = [22, 38, 67, 67]
list_graduated = ['No', 'Yes', 'Yes', 'Yes']
list_profession = ['Healthcare', 'Engineer', 'Engineer', 'Lawyer']
list_experience = [1.0, np.nan, 1.0, 0.0]
list_spending_score = ['Low', 'Average', 'Low', 'High']
list_family_size = [4.0, 3.0, 1.0, 2.0]
list_var_1 = ['Cat_4', 'Cat_4', 'Cat_6', 'Cat_6']
list_segmentation = ['D', 'A', 'B', 'B']
```

```python
# combining list with zip() function
list_zip = list(zip(list_id, list_gender, list_mariage,
                    list_age, list_graduated, list_profession,
                    list_experience, list_spending_score, list_family_size,
                    list_var_1, list_segmentation))
list_zip
```

```python
# creating a dataframe
df_list = pd.DataFrame(list_zip, columns=column_name)
df_list
```

### Solution - Task 2
Creating a dataframe using tuples

```python
# creating a tuple for each row
tuple_1 = (462809, 'Male', 'No',
           22, 'No', 'Healthcare',
           1.0, 'Low', 4.0,
           'Cat_4', 'D')

tuple_2 = (462643, 'Female', 'Yes',
           38, 'Yes', 'Engineer',
           np.nan, 'Average', 3.0,
           'Cat_4', 'A')

tuple_3 = (466315, 'Female', 'Yes',
           67, 'Yes', 'Engineer',
           1.0, 'Low', 1.0,
           'Cat_6', 'B')

tuple_4 = (461735, 'Male', 'Yes',
           67, 'Yes', 'Lawyer',
           0.0, 'High', 2.0,
           'Cat_6', 'B')

list_of_tuple = [tuple_1, tuple_2, tuple_3, tuple_4]
```

```python
# creating a dataframe
df_tuple = pd.DataFrame(list_of_tuple, columns=column_name)
df_tuple
```

### Solution - Task 4
Creating a dataframe using pandas series

```python
# creating a pandas series for each column
series_id = pd.Series([462809, 462643, 466315, 461735])
series_gender = pd.Series(['Male', 'Female', 'Female', 'Male'])
series_mariage = pd.Series(['No', 'Yes', 'Yes', 'Yes'])
series_age = pd.Series([22, 38, 67, 67])
series_graduated = pd.Series(['No', 'Yes', 'Yes', 'Yes'])
series_profession = pd.Series(['Healthcare', 'Engineer', 'Engineer', 'Lawyer'])
series_experience = pd.Series([1.0, np.nan, 1.0, 0.0])
series_spending_score = pd.Series(['Low', 'Average', 'Low', 'High'])
series_family_size = pd.Series([4.0, 3.0, 1.0, 2.0])
series_var_1 = pd.Series(['Cat_4', 'Cat_4', 'Cat_6', 'Cat_6'])
series_segmentation = pd.Series(['D', 'A', 'B', 'B'])

dict_of_series = {'ID':series_id,
                  'Gender':series_gender,
                  'Ever_Married':series_mariage,
                  'Age':series_age,
                  'Graduated':series_graduated,
                  'Profession':series_profession,
                  'Work_Experience':series_experience,
                  'Spending_Score':series_spending_score,
                  'Family_Size':series_family_size,
                  'Var_1':series_var_1,
                  'Segmentation':series_segmentation}
# dict_of_series
```

```python
# creating a dataframe
df_series = pd.DataFrame(dict_of_series)
df_series
```
