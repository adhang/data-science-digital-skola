# Hypothesis Testing

## Z-test
For this case, I will use `CO2` attribute (one sample case).

```python
overall_CO2_avg = data['CO2'].mean()
overall_CO2_std = data['CO2'].std()

print(f'The overall CO2 average is\t\t: {round(overall_CO2_avg,2)}')
print(f'The overall CO2 standard deviation is\t: {round(overall_CO2_std,2)}')
```

Our hypotheses:
- H0: The average CO2 produced when the transformer works is greater than the average CO2 produced (1816.41)
- H1: The average CO2 produced when the transformer works is less than the average CO2 produced (1816.41)
<br>

And suppose:
- We will use 35 sample using random sampling
- Confidence iterval is 95%, therefore alpha is 0.05 (5%)

```python
# stratified sampling
data_working = data[data['event'] == 'Working']

# random sampling 
sample_working_CO2 = data_working['CO2'].sample(n=35)

# z-test
ztest, pvalue = ztest(x1=sample_working_CO2, x2=None, value=overall_CO2_avg, alternative='larger')
print(f'ztest: {round(ztest,2)}, p-value: {round(pvalue,2)}')

# confidence interval and alpha
confidence_interval = 0.95
alpha = 1 - confidence_interval

# evaluating the hypothesis
print('\n')
if pvalue > alpha:
  print('Accept the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) > alpha ({round(alpha,2)})')
elif pvalue < alpha:
  print('Reject the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) < alpha ({round(alpha,2)})')
else:
  print('p-value and alpha is the same, decide yourself')
  print(f'Since p-value ({round(pvalue,2)}) = alpha ({round(alpha,2)})')
```

## T-test
For this case, I will use `Oxigen` attribute (one sample case). Yes, the column name is `Oxigen`, not `Oxygen` or `O2`. I just follow the dataset.

```python
overall_O2_avg = data['Oxigen'].mean()
overall_O2_std = data['Oxigen'].std()

print(f'The overall O2 average is\t\t: {round(overall_O2_avg,2)}')
print(f'The overall O2 standard deviation is\t: {round(overall_O2_std,2)}')
```

Our hypotheses:
- H0: The average O2 produced when the transformer works is less than the average O2 produced (8357.37)
- H1: The average O2 produced when the transformer works is greater than the average O2 produced (8357.37)
<br>

And suppose:
- We will use 25 sample using random sampling
- Confidence iterval is 95%, therefore alpha is 0.05 (5%)

```python
# stratified sampling
data_working = data[data['event'] == 'Working']

# random sampling 
sample_working_O2 = data_working['Oxigen'].sample(n=25)

# t-test
ttest, pvalue = stats.ttest_1samp(sample_working_O2, overall_O2_avg, alternative='less')
print(f'ttest: {round(ttest,2)}, p-value: {round(pvalue,2)}')

# confidence interval and alpha
confidence_interval = 0.95
alpha = 1 - confidence_interval

# evaluating the hypothesis
print('\n')
if pvalue > alpha:
  print('Accept the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) > alpha ({round(alpha,2)})')
elif pvalue < alpha:
  print('Reject the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) < alpha ({round(alpha,2)})')
else:
  print('p-value and alpha is the same, decide yourself')
  print(f'Since p-value ({round(pvalue,2)}) = alpha ({round(alpha,2)})')
```

## Chi-square
Suppose we want to know if there is a relationship between `CO`, `CO2`, `Hydrogen`, and `Methane` attributes based on the `event` attribute.
<br><br>
This is called chi-square test for independence, also called Pearson's chi-square test or the chi-square test of association. In this test we are testing the association between two or more variable ([source](https://stats.stackexchange.com/questions/110718/chi-squared-test-with-scipy-whats-the-difference-between-chi2-contingency-and)).
<br><br>
And suppose:
- Confidence iterval is 95%, therefore alpha is 0.05 (5%)

```python
chi = data[['event', 'CO', 'CO2', 'Hydrogen', 'Methane']]
chi = chi.set_index('event')
chi.head()
```

```python
# chi-square
chi2, pvalue, dof, expected = stats.chi2_contingency(chi)
print(f'chi2: {round(chi2,2)}, p-value: {round(pvalue,2)}')

# confidence interval and alpha
confidence_interval = 0.95
alpha = 1 - confidence_interval

# evaluating the hypothesis
print('\n')
if pvalue > alpha:
  print('Accept the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) > alpha ({round(alpha,2)})')
elif pvalue < alpha:
  print('Reject the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) < alpha ({round(alpha,2)})')
else:
  print('p-value and alpha is the same, decide yourself')
  print(f'Since p-value ({round(pvalue,2)}) = alpha ({round(alpha,2)})')
```

## Two Test Sample
For this case, I will use `Methane` attribute (two sample case).
- Case 1: The methane produced when the transformer is working
- Case 2: The methane produced when the transformer is not working

```python
overall_methane_avg = data['Methane'].mean()
overall_methane_std = data['Methane'].std()

print(f'The overall Methane average is\t\t\t: {round(overall_methane_avg,2)}')
print(f'The overall Methane standard deviation is\t: {round(overall_methane_std,2)}')
```

Our hypotheses:
- H0: The average methane produced when the transformer is working is the same as when the transformer is not working
- H1: The average methane produced when the transformer is working is not the same as when the transformer is not working

### Using Z-test
Suppose:
- We will use 50 sample using random sampling
- Confidence iterval is 95%, therefore alpha is 0.05 (5%)

```python
# stratified sampling
data_working = data[data['event'] == 'Working']
data_not_working = data[data['event'] == 'Not Working']

# random sampling 
sample_working_methane = data_working['Methane'].sample(n=50)
sample_not_working_methane = data_not_working['Methane'].sample(n=50)

# z-test
ztest, pvalue = ztest(x1=sample_working_methane, x2=sample_not_working_methane, value=0, alternative='two-sided')
print(f'ztest: {round(ztest,2)}, p-value: {round(pvalue,2)}')

# confidence interval and alpha
confidence_interval = 0.95
alpha = 1 - confidence_interval

# evaluating the hypothesis
print('\n')
if pvalue > alpha:
  print('Accept the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) > alpha ({round(alpha,2)})')
elif pvalue < alpha:
  print('Reject the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) < alpha ({round(alpha,2)})')
else:
  print('p-value and alpha is the same, decide yourself')
  print(f'Since p-value ({round(pvalue,2)}) = alpha ({round(alpha,2)})')
```

### Using T-test
Suppose:
- We will use 25 sample using random sampling
- Confidence iterval is 95%, therefore alpha is 0.05 (5%)

```python
# stratified sampling
data_working = data[data['event'] == 'Working']
data_not_working = data[data['event'] == 'Not Working']

# random sampling 
sample_working_methane = data_working['Methane'].sample(n=25)
sample_not_working_methane = data_not_working['Methane'].sample(n=25)

# t-test
ttest, pvalue = stats.ttest_ind(a=sample_working_methane, b=sample_not_working_methane, equal_var=True)
print(f'ttest: {round(ttest,2)}, p-value: {round(pvalue,2)}')

# confidence interval and alpha
confidence_interval = 0.95
alpha = 1 - confidence_interval

# evaluating the hypothesis
print('\n')
if pvalue > alpha:
  print('Accept the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) > alpha ({round(alpha,2)})')
elif pvalue < alpha:
  print('Reject the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) < alpha ({round(alpha,2)})')
else:
  print('p-value and alpha is the same, decide yourself')
  print(f'Since p-value ({round(pvalue,2)}) = alpha ({round(alpha,2)})')
```
