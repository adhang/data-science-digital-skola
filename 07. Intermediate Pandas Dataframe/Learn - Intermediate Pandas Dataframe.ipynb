# Sorting Dataframe

## Sort Dataframe by Specific Column
Syntax
```
dataframe.sort_values(by='column_name')
```

By default, sorting will be done in ascending order.

````python
# sort dataframe by employee_id column
data_10.sort_values(by='employee_id')
````

We can change the sorting order by setting the `ascending` parameter.

````python
# sort in descending order
data_10.sort_values(by='employee_id', ascending=False)
````

## Sort Single Column
We can take a specific column and then sort it. The outpus will be a series.

````python
# get employee_id column and sort it
data_10['employee_id'].sort_values()
````

````python
# sort in descending order
data_10['employee_id'].sort_values(ascending=False)
````

## Sort by Multiple Columns
Sort by first column, then the second column.

````python
data_10.sort_values(by=['department','age'])
````

Sort the data by department, then we sort by age. You can see at the first 2 rows have the same `department` name. Therefore, the order of those first 2 rows is determined by the `age` column.

# Filtering Dataframe
Let's create a new dataframe using random sampling.

````python
data_20 = data.sample(n=20)
data_20.shape
````

## Get Specific Columns
Getting some specific columns in Pandas is similar to the `SELECT` statement in SQL. For example, we want to get `department` and `awards_won` columns.

### Using Square Brackets
Use double square brackets `[[]]` to create a dataframe. If we only use a single bracket `[]`, it will create a series. It's a native way to create a series or a dataframe.

````python
data_20[['department','awards_won']].head()
````

### Using Function
Pandas has a built-in method to select specific columns.

````python
data_20.filter(items=['department','awards_won']).head()
````

## Data Selection
- `loc` - used to select rows or columns by its name
- `iloc` - used to select rows or columns by its index

I can say, `loc` stands for `location`, and `iloc` stands for `integer location` or `index location`. Just to make it easier to memorize the difference.

But first, I'm going to reset the index of my dataframe.

````python
data_20 = data.sample(n=20)
data_20.reset_index(inplace=True)
data_20.head()
````

You can see that the index from the dataframe becomes a new column. We don't want this. So, We have to use the `drop` parameter when resetting the index.

````python
data_20 = data.sample(n=20)
data_20.reset_index(inplace=True, drop=True)
data_20.head()
````

### loc
Let's say, we want to get the first 5 rows for the `is_promoted` and `gender` columns. Since we will use column names, we have to use the `loc` method.

````python
data_20.loc[3:5, ['is_promoted', 'gender']]
````

### iloc
Let's see the last 10 rows using `tail()`.

````python
data_20.tail(10)
````

For this case, we want to get the last 10 rows for the `recruitment_channel` and `no_of_trainings` columns. But this time, we will use the index values.

Just like a Numpy array, the index of rows is started from 0, and the index of columns is started from 0. So, if we can say that `recruitment_channel` is located in the 6th column (index = 5) and `no_of_trainings` column is located in the 7th column (index = 6). Remember, the upper-bound of slicing is exclusive (not included).

````python
data_20.iloc[-10:, 5:7]
````

### Select By Conditions
What if we want to filter our data using a condition? Let's say we want to get data where the `age` is above 35.

````python
filter_data = data_20.loc[data_20['age'] > 35]
filter_data
````

As you can see, we get new dataframe where the `age` is above 35. How about using multiple conditions? For example, we want to get data where the `age` is above 35 and the `length_of_service` is above 4.

````python
filter_data = data_20.loc[(data_20['age'] > 35) & (data_20['length_of_service'] > 4)]
filter_data
````

We have filtered our data using 2 conditions. Filtering data using conditions in Pandas is similar to the `WHERE` statement in SQL.

Then, how about selecting specific columns and filtering the data at the same time?

This time, I want to get data where the `age` is above 35 and the `lenght_of_service` is above 4. Beside that, I only want to display `employee_id`, `department`, `region`, `age`, and `length_of_service` columns.

````python
filter_data = data_20.loc[(data_20['age'] > 35) & (data_20['length_of_service'] > 4), ['employee_id','department','region','age','length_of_service']]
filter_data
````

The code below is another way to do the same thing as the above code.

````python
filter_data = data_20[['employee_id','department','region','age','length_of_service']].loc[(data_20['age'] > 35) & (data_20['length_of_service'] > 4)]
filter_data
````

Sometimes, we need to create some new columns. For example, we have a data like this:

| **id** | **name** | **quiz** | **homework** |
|:------:|:--------:|:--------:|:------------:|
|    1   |  Adhang  |    80    |      85      |

We have `quiz` and `homework` scores. Then, we need to create a new column to store the `total` score. We can do it using Pandas.

For this dataset, I will use a dummy calculation that doesn't have any meaning (or usage).

````python
# sum 'age' and 'length_of_service'
filter_data['dummy_sum'] = filter_data['age'] + filter_data['length_of_service']
filter_data.head()
````

A new column named `dummy_sum` has been created.

Can we concatenate the string of some columns? Yes we can! We have to use `astype()` to cast a data type.

````python
filter_data['str_sum'] = filter_data['age'].astype(str) + filter_data['length_of_service'].astype(str)
filter_data.head()
````

Now, we have another new column named `str_sum` containing concatenated string values of the `age` and the `length_of_service` columns.

# Grouping Dataframe
Pandas has similar method to `GROUP BY` statement in SQL. Grouping means, we will group our data based on specific columns and doing some calculation (aggregation).

## Group by Single Column

### Aggregating All Columns
For example, we want to calculate all columns' average and group it by the `department` column.

````python
filter_data.groupby('department').mean()
````

We have a new dataframe that contains aggregation result.

### Aggregating Specific Columns
What if we only want to know the average age of each department? We can select the `department` and the `age` columns, then group by `department`.

````python
filter_data[['department','age']].groupby('department').mean()
````

### Multiple Aggregations
The above example only use `mean()` to calculate the average of a column. But, can we use different aggregation method?

We can use `agg()` to do different aggregations.

Syntax


```
dataframe.groupby('column_name').agg(new_col_1 = ('col_name_1', 'agg_method_1'),
                                    (new_col_2 = ('col_name_2', 'agg_method_2'))
```

How to calculate the average age and maximum length of service of each department?


````python
filter_data.groupby('department').agg(avg_age = ('age', 'mean'),
                                      max_services = ('length_of_service', 'max'))
````

We got a new dataframe with the `department` column becomes the index label. If you don't want to set `department` as index, you can use `reset_index()`.

````python
filter_data.groupby('department').agg(avg_age = ('age', 'mean'),
                                      max_services = ('length_of_service', 'max')).reset_index()
````

## Group by Multiple Columns
Just like sorting, we can group our data using multiple columns.

Let's make a new dataframe containing 100 randomly sampled data.

````python
data_100 = data.sample(n=100)
data_100.head()
````

I want to know the average age of each department and education.

````python
data_100[['department','education','age']].groupby(['department','education']).mean().round(2)
````

# Merging Dataframe
Just like SQL JOINs. It used to combine several tables in terms of columns (combining both tables' columns).

Syntax

```
left_table.merge(right_table, how='join_type', on='column_name')
```

To combine 2 tables, we need a connector that will be placed on `on` parameter. In database term, we need a primary key and a foreign key. In this dataset, the `employee_id` acts as the connector.

````python
data.head()
````

I will create 2 dataset from a single dataset
- The first dataset will be the left table. It contains the `employee_id`, `department`, `region`, `education`, `gender`, and `recruitment_channel` columns.

- The second dataset will be the right table. it contains the `employee_id`,	`is_promoted`, and	`no_of_trainings` columns.

````python
data_left = data.iloc[:20, :6]
data_left.head()
````

````python
data_right = data.loc[18:30, ['employee_id', 'is_promoted', 'no_of_trainings']]
data_right.head()
````

## Inner Join
The output is only the intersected data (`B`), the data that appears in both table `A` and table `C`.

<img src='https://raw.githubusercontent.com/adhang/data-science-digitalskola/main/07.%20Intermediate%20Dataframe/res/inner_join.png' width=300>

````python
data_left.merge(data_right, how='inner', on='employee_id')
````

There's an intercept between table `A` and table `C` where the `employee_id` is 17436 and 54461. In other words, these two `employee_id` records appear on both tables.

## Left Join
The output is only the intersected data (`B`) and all data that appears on table `A`. The data records on table `A` that do not intersect with/ appear on table `C` will be `NaN`.

<img src='https://raw.githubusercontent.com/adhang/data-science-digitalskola/main/07.%20Intermediate%20Dataframe/res/left_join.png' width=300>

````python
data_left.merge(data_right, how='left', on='employee_id')
````

The `is_promoted` and `no_of_trainings` columns are column from table `C`. On these columns, the records from index 0-17 are `NaN` because they only appear on table `A` but not in table `C`. You can see that the last two rows (index 18 and 19) are the same as the result of the inner join.

## Right Join
The output is only the intersected data (`B`) and all data that appears on table `C`. The data records on table `C` that do not intersect with table `A` will be `NaN`.

<img src='https://raw.githubusercontent.com/adhang/data-science-digitalskola/main/07.%20Intermediate%20Dataframe/res/right_join.png' width=300>

````python
data_left.merge(data_right, how='right', on='employee_id')
````

The `department`, `region`, `education`, `gender`, and `recruitment_channel` columns are column from table `A`. On these columns, the records from index 2-12 are `NaN` because they only appear on table `C` but not in table `A`. You can see that the first two rows (index 0 and 1) are the same as the result of the inner join.

## Outer Join
The output is the combination of all data from table `A` and table `C`. The data records that do not intersect will be `NaN`.

<img src='https://raw.githubusercontent.com/adhang/data-science-digitalskola/main/07.%20Intermediate%20Dataframe/res/outer_join.png' width=300>

````python
data_left.merge(data_right, how='outer', on='employee_id')
````

From this result, we know that the records with index 0-17 are the records that only appear on table `A`, so they have `NaN` values on the `is_promoted` and `no_of_trainings` columns.

On the other hand, the records from index 20-30 are the records that only appear on table `C`, so they have `NaN` values on the `department`, `region`, `education`, `gender`, and `recruitment_channel` columns.

The rows with index 18 and 19 have complete records on all columns, that's because the `employee_id` values exist on both tables.
