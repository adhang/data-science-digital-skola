```python
sql_script =  """
              with 
                t1 as
                (
                  select *
                  from batch_11.cb_trips
                )
              
              select *
              from t1
              """

data_trips = pd.read_sql_query(sql_script, pg_connection)
data_trips.head()
```

Join cb_stations with cb_trips on station_id (cb_stations) = end_station_id (cb_trips)

```python
sql_script =  """
              with 
                t1 as
                (
                  select
                    cs.name,
                    ct.tripduration
                  from batch_11.cb_trips ct
                  join batch_11.cb_stations cs
                  on cs.station_id = ct.end_station_id
                )
              
              select
                t1.name,
                sum(t1.tripduration)
              from t1
              where lower(t1.name) like '%clermont%'
              group by 1
              having sum(t1.tripduration) < 3000000
              """

data_trips = pd.read_sql_query(sql_script, pg_connection)
data_trips.head()
```

```python
sql_script =  """
              select *
              from batch_11.cb_trips ct
              where ct.gender = 0 and ct.tripduration = 634
              """

data_challenge_1 = pd.read_sql_query(sql_script, pg_connection)
data_challenge_1.head()
```

```python
station = data_challenge_1['start_station_name'].value_counts()
print(station)
```

```python
sql_script =  """
              select *
              from batch_11.cb_trips ct
              where ct.gender = 0 and ct.tripduration = 634
              """

data_challenge_1 = pd.read_sql_query(sql_script, pg_connection)
# data_challenge_1.head()

station = data_challenge_1['start_station_name'].value_counts()

df_station = pd.DataFrame(station)
df_station.reset_index(inplace=True)
df_station.rename(columns={'index':'start_station', 'start_station_name':'total'}, inplace=True)
df_station
```
