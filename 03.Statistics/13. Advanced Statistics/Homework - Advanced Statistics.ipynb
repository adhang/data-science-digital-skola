# 4. Hypothesis Testing
In this dataset, the `credit_limit` attribute is one of the variables to determine whether the customer will stop (churn/ attrition) or continue to use the product. I will do a hypothesis testing whether the average credit limit for men and women is the same or not.
<br>

Hypothesis:
- H0 = the average credit limit for men and women is the same
- H1 = the average credit limit for men and women is not the same

Hypotesis testing method:
- Using 50 samples
- The confidence interval is 95% (0.95)
- Since this hypothesis is used to compare the average value between category (men and women), we will use two sample test
- Since we know the variance of population and use 50 samples, we will use z-test to test this hypothesis

```python
overall_credit_limit_avg = data['credit_limit'].mean()
overall_credit_limit_std = data['credit_limit'].std()

print(f'The overall credit limit average is\t\t: {round(overall_credit_limit_avg,2)}')
print(f'The overall credit limit standard deviation is\t: {round(overall_credit_limit_std,2)}')
```

```python
# stratified sampling
data_female = data[data['gender'] == 'F']
data_male = data[data['gender'] == 'M']

# random sampling 
sample_data_female = data_female['credit_limit'].sample(n=50)
sample_data_male = data_male['credit_limit'].sample(n=50)

# z-test
ztest, pvalue = ztest(x1=sample_data_female, x2=sample_data_male, value=0, alternative='two-sided')
print(f'ztest: {round(ztest,2)}, p-value: {round(pvalue,2)}')

# confidence interval and alpha
confidence_interval = 0.95
alpha = 1 - confidence_interval

# evaluating the hypothesis
print('\n')
if pvalue < alpha:
  print('Reject the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) < alpha ({round(alpha,2)})')
else:
  print('Accept the hypothesis')
  print(f'Since p-value ({round(pvalue,2)}) > alpha ({round(alpha,2)})')
```

Since p-value is less than alpha, we reject the null hypothesis (H0). Hence, we accept the alternative hypothesis (H1).
<br><br>
In conclusion, the average credit limit for men and women is not the same.
