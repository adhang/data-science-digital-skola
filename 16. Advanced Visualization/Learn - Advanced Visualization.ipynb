# Seaborn

## Importing Library

```python
import seaborn as sns
import matplotlib.pyplot as plt

# set plot to sns theme, even if we use matplotlib
sns.set_theme(style='darkgrid')

# another way to set the style:
# sns.set_style('darkgrid')
```

## Reading Dataset

```python
data = sns.load_dataset('tips')
data.head()
```

- `total_bill` - total charged (USD)
- `tip` - tips (USD)
- `sex` - gender (Female, Male)
- `smoker` - smoker or not (Yes, No)
- `day` - day (Sun, Sat, Thur, Fri)
- `size` - number of people in visitor group

```python
data.describe()
```

```python
data.info()
```

```python
data.isna().sum()
```

## Line Plot

### Basic Line Plot

```python
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

fig, ax = plt.subplots(figsize=(9,6))

sns.lineplot(x=xpoints, y=ypoints, ax=ax)

plt.show()
```

### Regression Plot

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.regplot(data=data, x='total_bill', y='tip', ax=ax)

# plt.show()
```

```python
# same as regplot but using FacetGrid
sns.lmplot(data=data, x='total_bill', y='tip')

# plt.show()
```

```python
# same as regplot but using FacetGrid
sns.lmplot(data=data, x='total_bill', y='tip', hue='smoker')

# plt.show()
```

```python
# same as regplot but using FacetGrid
sns.lmplot(data=data, x='total_bill', y='tip', hue='time')

# plt.show()
```

```python
# same as regplot but using FacetGrid
sns.lmplot(data=data, x='total_bill', y='tip', hue='sex')

# plt.show()
```

## Scatter Plot

### Basic Scatter Plot

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.scatterplot(data=data, x='total_bill', y='tip', hue='sex', ax=ax)

plt.show()
```

```python
sns.lmplot(data=data, x='total_bill', y='tip', hue='sex', fit_reg=False)
```

### Bubble Chart

```python
fig, ax = plt.subplots(figsize=(9,6))

sns.scatterplot(data=data, x='total_bill', y='tip', hue='sex', size='size', ax=ax)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
