### Solution - Main Task
Merging datasets with outer join

```python
data_merge = data_1.merge(data_2, how='outer', on='employee_id')
data_merge.head()
```

The column names contain `_x` or `_y` postfix, except the `employee_id` column. That's because, we are merging (joining - in SQL term) 2 tables with `employee_id` as a connector, which means that the columns are considered as the same column. The other columns are considered as different columns, even the column names are the same. There is a reason behind it. Let's see the 2 tables below.

> **order_table**

| **order_id** | **user_id** |  **name**  | **price** |
|:------------:|:-----------:|:----------:|:---------:|
|       1      |      2      |    mango   |    4000   |
|       2      |      1      | strawberry |    5000   |
|       3      |      1      |    lemon   |    4500   |

> **user_table**

| **user_id** | **name** |  **city** |
|:-----------:|:--------:|:---------:|
|      1      |  Adhang  |   Jogja   |
|      2      |  Muntaha |  Jakarta  |
|      3      | Muhammad | Palembang |

Both tables contain a `name` column. What if we merge those tables and assume that the `name` column is the same for both tables? It will be a disaster because the `name` column on `order_table` represents the product names, but the `name` column on `user_table` represents the user names.
<br><br>
To handle this problem, Pandas is assuming that they are different column even the column name are the same. So, Pandas automatically add a postfix to represent it as a different column.

```python
data_merge.info()
```

### Solution - Task 1
Filter the data based on the `awards_won_y` value. I will only get the if the `awards_won_y` is null. That's because, when we merge the tables:
- the columns from `data.csv` have `_x` as the postfix 
- the columns from `data_2.csv` have `_y` as the postfix

That means the first dataset will store its datapoint on a column with `_x` postfix. Then, the first dataset will have `NaN` values for every column that has `_y` postfix.

```python
data_filter = data_merge[data_merge['awards_won_y'].isnull()]
data_filter.head()
```

### Solution - Task 3
The column that has `_x` postfix is the first column (index = 0) until the 14th column (index = 13). We can filter it based on the column index. I use `:14` because the upper-bound of slicing is exclusive (not included).

```python
data_slice = data_filter.iloc[:,:14]
data_slice.head()
```

```python
data_slice.shape
```

Voila! The dataframe has changed to have 14 columns.

### Solution - Task 4
To rename the column name, I will make a list containing the first dataset column names.

```python
column_list = list(data_1.columns)
column_list
```

I'm creating the merged dataset column names, just by adding `_x` postfix.

```python
column_list_x = []

for i in range(0, len(column_list)):
  column_list_x.append(column_list[i] + '_x')

column_list_x
```

Now, let's rename it.

```python
for i in range(0,len(column_list)):
  data_slice.rename(columns={column_list_x[i]:column_list[i]}, inplace=True)

data_slice.head()
```

Finally, the dataframe column names have changed!
